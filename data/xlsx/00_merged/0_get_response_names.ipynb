```python
import os
import glob
import re
import numpy as np
import pandas as pd

# Folder containing the Excel files
folder = r"C:\Users\david\OneDrive - Univerzita Karlova\GAUK Michal\parsing piratske forum\data\xlsx\00_merged"

# Regex pattern: token.token before ' píše:'
pattern = r'([A-Za-zÁČĎÉĚÍŇÓŘŠŤÚŮÝŽáčďéěíňóřšťúůýž]+(?:\.[A-Za-zÁČĎÉĚÍŇÓŘŠŤÚŮÝŽáčďéěíňóřšťúůýž]+)+)\s+píše:'

def extract_unique(names):
    seen = set()
    unique = []
    for name in names:
        if name not in seen:
            seen.add(name)
            unique.append(name)
    return unique

# Loop over all Excel files
for file in glob.glob(os.path.join(folder, "*.xlsx")):
    print(f"Processing: {file}")

    df = pd.read_excel(file)

    # Create response column
    df['response'] = (
        df['content']
          .astype(str)
          .str.findall(pattern, flags=re.UNICODE)
          .apply(lambda names: ", ".join(extract_unique(names)) if names else np.nan)
    )

    # Save back to the same file
    df.to_excel(file, index=False)

print("✅ All files processed and saved.")
```
