```python
import pandas as pd
from pathlib import Path

# Folder containing the xlsx files
folder_path = Path(r"C:\Users\david\OneDrive - Univerzita Karlova\GAUK Michal\parsing piratske forum\Political-Party-Elites\data\xlsx\00_merged")

# Output file path
output_file = folder_path / "merged_all.xlsx"

# Find all xlsx files (excluding the output file if it already exists)
xlsx_files = [f for f in folder_path.glob("*.xlsx") if f.name != output_file.name]

# Read and merge all files
dfs = []
for file in xlsx_files:
    df = pd.read_excel(file)
    df["source_file"] = file.name  # optional: keeps track of origin
    dfs.append(df)

merged_df = pd.concat(dfs, ignore_index=True)

# Save merged file
merged_df.to_excel(output_file, index=False)

print(f"Merged {len(xlsx_files)} files into: {output_file}")
```
